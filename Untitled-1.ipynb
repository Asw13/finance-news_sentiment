```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from transformers import TFDistilBertModel
import tensorflow as tf
from keras.saving import register_keras_serializable
from transformers import DistilBertTokenizer
import subprocess
import pandas as pd
```

```python


class TransformerClassifier(Model):
    def __init__(self, **kwargs):
        super(TransformerClassifier, self).__init__(**kwargs)
        self.bert = TFDistilBertModel.from_pretrained("distilbert-base-uncased")
        self.bn = BatchNormalization()
        self.dropout = Dropout(0.2)
        self.fc = Dense(256, activation="relu")
        self.output_layer = Dense(3, activation="softmax")

    def call(self, inputs):
        bert_output = self.bert(inputs)[0][:, 0, :]  # CLS token output
        x = self.bn(bert_output)
        x = self.dropout(x)
        x = self.fc(x)
        return self.output_layer(x)

    def get_config(self):
        config = super().get_config()
        # You can also add any custom attributes if needed
        return config

    @classmethod
    def from_config(cls, config):
        return cls(**config)
```

# **Load the Sentiment Model**

```python


new_model = tf.keras.models.load_model(
    "distilbert_full_model.keras",
    custom_objects={"TransformerClassifier": TransformerClassifier}
)

loaded_tokenizer = DistilBertTokenizer.from_pretrained("DistilBERT_saved_tokenizer")
```

# **Predict Sentiment**

```python

# Define class labels
class_labels = {0: "Negative", 1: "Neutral", 2: "Positive"}


# Function to predict sentiment
def predict_sentiment(text):
    encoding = loaded_tokenizer(text, truncation=True, padding=True, return_tensors="tf")
    prediction = new_model.predict({"input_ids": encoding["input_ids"], "attention_mask": encoding["attention_mask"]})
    predicted_class = tf.argmax(prediction, axis=1).numpy()[0]
    return class_labels[predicted_class]

# Apply sentiment prediction on the 'News' column
df["Sentiment"] = df["News"].apply(predict_sentiment)

# Display the updated dataframe
print(df.head())
```

# **Group and Aggregate**

```python

# Ensure date is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Function to determine the grouping date
def get_group_date(timestamp):
    weekday = timestamp.weekday()
    hour = timestamp.hour
    if weekday == 4 and hour >= 16:
        return (timestamp + pd.DateOffset(days=3)).normalize()
    elif weekday == 5 or weekday == 6:
        return (timestamp + pd.DateOffset(days=(7 - weekday))).normalize()
    else:
        return timestamp.normalize()

# Apply the grouping logic
df['Group_Date'] = df['Date'].apply(get_group_date)

# Group by 'Group_Date' and 'Sentiment' and count occurrences
sentiment_counts = df.groupby(['Group_Date', 'Sentiment']).size().unstack(fill_value=0)

# Rename columns if needed
sentiment_counts = sentiment_counts.rename(columns={
    'Positive': 'Positive',
    'Neutral': 'Neutral',
    'Negative': 'Negative'
}).reset_index()

# If any sentiment label is missing in the data, ensure those columns exist
for sentiment in ['Positive', 'Neutral', 'Negative']:
    if sentiment not in sentiment_counts.columns:
        sentiment_counts[sentiment] = 0

# Final grouped DataFrame
print(sentiment_counts)
```

# **Visualize**

```python
import matplotlib.pyplot as plt

# Make sure 'Group_Date' is datetime
sentiment_counts['Group_Date'] = pd.to_datetime(sentiment_counts['Group_Date'])

# Set figure size
plt.figure(figsize=(12, 6))

# Plot stacked bars
plt.bar(sentiment_counts['Group_Date'], sentiment_counts['Positive'], label='Positive', color='green')
plt.bar(sentiment_counts['Group_Date'], sentiment_counts['Neutral'], bottom=sentiment_counts['Positive'], label='Neutral', color='gray')
plt.bar(sentiment_counts['Group_Date'],
        sentiment_counts['Negative'],
        bottom=sentiment_counts['Positive'] + sentiment_counts['Neutral'],
        label='Negative', color='red')

# Format the plot
plt.xlabel('Group Date')
plt.ylabel('Sentiment Count')
plt.title('Sentiment Distribution Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```
